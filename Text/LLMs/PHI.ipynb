```python
import pandas as pd


df=pd.read_csv(r"C:\Users\omkar\OneDrive\Documents\Internship\SA\LLMs\Phi_imdb.csv")
```

```python
df['pred_label'] = df['pred_label'].replace({'LABEL_1':'positive','LABEL_0' :'negative'})
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
import matplotlib.pyplot as plt
```

```python
print(classification_report(df['sentiment'],df['pred_label']))
cm = confusion_matrix(df['sentiment'],df['pred_label'], labels=["negative", "positive" ])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["negative", "positive"])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
